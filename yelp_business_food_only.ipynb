```python
#분석에 쓰일 유용한 함수 클래스 생성
#클래스 안의 unique 함수에 에러처리 포함
```

```python
class UsefulFunc():
    def __init__(self):
        self._new = 0
    
    def checktime(self, func):
        def new_func(*args, **kwargs):
            timestamp = datetime.datetime.now().strftime("%Y/%m/%d %H:%M")
            print("[{}]".format(timestamp))
            start = time.time()
            result = func(*args, **kwargs)
            #time.sleep(2)
            end = time.time()
            print("\n실행시간: ", end - start, "초", sep = '')
            return result
        return new_func

    def unique(self, lst):
        try:
            self._new = []
            for i in lst:
                if i not in self._new:
                    self._new.append(i)
        except:
            print("반복자아님")
        
        return self._new
```

```python
cat_lst = tor_food.split("\n") + tor_rest.split("\n") + edin_food.split("\n") + edin_rest.split("\n")
cat_lst
```

```python
#UsefulFunc 클래스에 있는 unique함수를 이용해, 
#리스트를 중복제거를 해준후 줄바꿈(\n)을 |(or)로 다 바꿔줌
```

```python
cat_str = '|'.join(b.unique(cat_lst))
cat_str
```

```python
#정규표현식을 통해, 찾은 카테고리를 통해서 
#컴파일시켜 전부찾게 하여 해당되는 카테고리 전부 뽑아오기
```

```python
lst = []
for i in business.categories:
    p = re.compile(r''+cat_str+'')
    if p.findall(i) == []:
        lst.append(0)
    else:
        lst.append(i)
```

```python
len(lst)
```

```python
lst_num = []
for i,v in enumerate(lst):
    if v != 0: lst_num.append(i)
```
