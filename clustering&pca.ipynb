```python
#분석에 쓰일 유용한 함수 클래스 생성
#클래스 안의 unique 함수에 에러처리 포함
```

```python
class UsefulFunc():
    def __init__(self):
        self._new = 0
    
    def checktime(self, func):
        def new_func(*args, **kwargs):
            timestamp = datetime.datetime.now().strftime("%Y/%m/%d %H:%M")
            print("[{}]".format(timestamp))
            start = time.time()
            result = func(*args, **kwargs)
            #time.sleep(2)
            end = time.time()
            print("\n실행시간: ", end - start, "초", sep = '')
            return result
        return new_func

    def unique(self, lst):
        try:
            self._new = []
            for i in lst:
                if i not in self._new:
                    self._new.append(i)
        except:
            print("반복자아님")
        
        return self._new
```

```python
from sklearn.preprocessing import StandardScaler
X_tor = StandardScaler().fit_transform(user_tor[["review_count","fans","friends_num","elite_num"]])
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

pca = PCA()
pipeline = make_pipeline(pca)
pipeline.fit(X_tor)

features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.show()
```

```python
pca = PCA(n_components = 4)
pc = pca.fit_transform(X_tor)

cum_explained_var = []
for i in range(0, len(pca.explained_variance_ratio_)):
    if i == 0:
        cum_explained_var.append(pca.explained_variance_ratio_[i])
    else:
        cum_explained_var.append(pca.explained_variance_ratio_[i] + 
                                 cum_explained_var[i-1])

print(cum_explained_var)
```

```python
#K-MEANS 분석
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

distortions = []
for k in range(1,10):
    kmeanModel = KMeans(n_clusters=k).fit(X_tor)
    kmeanModel.fit(X_tor)
    distortions.append(sum(np.min(cdist(X_tor, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X_tor.shape[0])
# Plot the elbow
plt.plot(range(1,10), distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
from sklearn.cluster import KMeans

num_clusters = 4

km = KMeans(n_clusters=num_clusters, init = "random")
km.fit(X_tor)
```

```python
df_tor = pd.DataFrame(X_tor, columns = ["review_count","fans","friends_num","elite_num"])
df_tor['cluster'] = km.labels_
```

```python
df_tor_desc = df_tor.groupby("cluster")["review_count", "fans", "friends_num", "elite_num"].mean()
```

```python
df_tor_desc["size"] = [len(df_tor[df_tor['cluster']==0]),len(df_tor[df_tor['cluster']==1]),len(df_tor[df_tor['cluster']==2]),len(df_tor[df_tor['cluster']==3])]
df_tor_desc
```
